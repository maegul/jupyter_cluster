```python
# All female abalone
female_mask = (abalone['sex'] == "F")

abalone.loc[female_mask]
```

```python
ring_mask = (abalone['rings'] >= 15)
abalone.loc[ring_mask]
```

```python
female_ring_mask = female_mask & ring_mask   # Elementwise AND - both must be True

abalone.loc[female_ring_mask]
```

```python
female_or_ring_mask = female_mask | ring_mask   # Elementwise OR - either can be True

abalone.loc[female_or_ring_mask]
```

```python
not_females_mask = ~female_mask  # Elementwise NOT - negates original
abalone.loc[not_females_mask]
```

```python
def extract_manufacturer(car_name):
    try:
        manufacturer, other = car_name.split(maxsplit=1)
    except ValueError as e:
#         print(f"Error with {car_name}: {e}")
        manufacturer = car_name
    return manufacturer
```
